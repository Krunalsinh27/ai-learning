```python
import numpy as np
```

NUMPY INDEXING AND SLICING OF VECTOR

```python
arr = np.arange(11,21)
arr
```

```python
arr_index = arr[5]
arr_index
```

```python
arr_slice = arr[3:9]
arr_slice
```

```python
arr_slice1 = arr[:7]
arr_slice1
```

```python
arr_slice2 = arr[3:]
arr_slice2
```

```python
arr_slice3 = arr[2:9:2]
arr_slice3
```

NUMPY INDEXING AND SLICING OF MATRIX

```python
arr = np.arange(1,31).reshape(5,6)
arr                              
```

```python
arr_index = arr[3,5]
arr_index
```

```python
slice = arr[2:5,4:6]
slice
```

```python
arr[:,3]
```

BOOLEAN INDEXING

```python
arr = np.arange(11,31)
arr
```

```python
bool_index = arr % 2 == 0
bool_index
```

```python
arr1 = arr[bool_index]
arr1
```

```python
import numpy as np
import pandas as pd
```

**Merging 2 Dataframes**

```python
employees = pd.DataFrame({
    'Emp_id': [1,2,3,4,5],
    'Name': ['Krunalsinh', 'Raj', 'Ram', 'Shyam', 'Bharat'],
    'Department': ['HR', 'IT', 'CE', 'CIVIL', 'EC']
})

salaries = pd.DataFrame({
    'Emp_id': [1,2,3,6,7],
    'Salary': [60000, 30000, 45000, 50475, 23000],
    'Bonus': [5000, 2000, 12000, 1000, 2200]
})
```

```python
employees
```

```python
salaries
```

```python
pd.merge(employees, salaries)
```

```python
pd.merge(employees, salaries, on='Emp_id')
```

```python
pd.merge(employees, salaries, on='Emp_id', how='outer')
```

```python
pd.merge(employees, salaries, on='Emp_id', how='left')
```

```python
pd.merge(employees, salaries, on='Emp_id', how='right')
```

**Concatination of 2 Dafaframes**

```python
df1 = pd.DataFrame({
    'A':['A0','A1','A2'],
    'B':['B0','B1','B2'],
    'C':['C0','C1','C2']
})

df2 = pd.DataFrame({
    'A':['A3','A4','A5'],
    'B':['B3','B4','B5'],
    'C':['C3','C4','C5']
})
```

```python
pd.concat([df1, df2])
```

```python
pd.concat([df2, df1])
```

```python
pd.concat([df1,df2], axis=1)
```

```python
pd.concat([df2,df1], axis=1)
```

**Joining of 2 Dafaframes**

```python
df1 = pd.DataFrame({
    'Name':['Krunalsinh', 'Raj', 'Shyam']
}, index=[1,2,3])

df2 = pd.DataFrame({
    'Runs':[102, 42, 53]
}, index=[2,3,4])
```

```python
df1
```

```python
df2
```

```python
df1.join(df2)
```

```python
df1.join(df2, how='outer')
```

```python
df2.join(df1)
```

```python
df2.join(df1, how='left')
```

```python
df2.join(df1, how='right')
```

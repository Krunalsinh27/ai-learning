```python
import numpy as np
import pandas as pd
```

**GroupBy**

```python
data = {
    'Category': ['A', 'B', 'A', 'B', 'A', 'B', 'A', 'B'],
    'Store': ['S1', 'S1', 'S2', 'S2', 'S1', 'S2', 'S2', 'S1'],
    'Sales': [100, 200, 150, 250, 120, 180, 200, 300],
    'Quantity': [10, 15, 12, 18, 8, 20, 15, 25],
    'Date': pd.date_range('2026-06-29', periods=8)
}
df = pd.DataFrame(data)
```

```python
df
```

```python
#Group by Category and Calculate the sum of Sales
cat = df.groupby('Category')
for i,v in cat:
    print(i)
    print(v)
```

```python
cat = df.groupby('Category')['Sales'].sum()
cat
```

```python
#Group by store and Calculate the sum of Sales
cat1 = df.groupby('Store')
for i,v in cat1:
    print(i)
    print(v)
```

```python
cat1 = df.groupby('Store')['Sales'].sum()
cat1
```

```python
# Group by multiple Columns
# Group by Category and Store
cat2 = df.groupby(['Category','Store'])
for i,v in cat2:
    print(i)
    print(v)
```

```python
cat2 = df.groupby(['Category','Store'])['Quantity'].sum()
cat2
```

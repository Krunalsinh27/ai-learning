PANDAS SERIES

Creating Series

```python
import numpy as np
import pandas as pd
```

```python
labels = ['a', 'b', 'c']
my_lists = [10, 20, 50]
arr = np.array([20,49,1])
my_disc = {'name':'krunalsinh', 'age':21, 'surname':'Parmar'}
```

```python
pd.Series(my_lists)
```

```python
pd.Series(my_lists, index=labels)
```

```python
pd.Series(arr)
```

```python
pd.Series(my_disc)
```

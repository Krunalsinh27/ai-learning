```python
import numpy as np
import pandas as pd
```

Finding Missing Data

```python
data = {
    'A':[1, 3, np.nan, 5, 6],
    'B':[np.nan, 5, 7,3,8],
    'C':[2, 6,8,np.nan, 2],
    'D':[7,9,np.nan, 4,np.nan]
}
df = pd.DataFrame(data)
```

```python
df
```

```python
df.isna()
```

```python
df.isna().sum()
```

```python
df.isna().any()
```

Removing Missing Data

```python
df
```

```python
df.dropna()
```

```python
df
```

```python
df.dropna(thresh=3)
```

Filling the Missing Data

```python
df.fillna(0)
```

```python
df.fillna(1)
```

```python
values = {'A':1, 'B':10, 'C':20, 'D':30}
df.fillna(value=values)
```

```python
df
```

```python
df.fillna(df.mean())
```
